```python
import os
import json 
import numpy as np
import pandas as pd
from ephys_template_data_preprocess_utils import process_ephys_subthresh_data, process_ephys_rheobase_data

```

```python
data_df = pd.read_csv(r'\\allen\programs\celltypes\workgroups\mousecelltypes\SarahWB\datasets\human_exc\data\ephys\20250603\human_excite_cells.csv')

#HANN mapping
subclass_col = 'HANN_subclass_label_assignment_winner'
cluster_col = 'HANN_cluster_label_assignment_winner'
```

# Process subthreshold data

```python
# process subthreshold data and save
subthresh_out_dir = r'\\allen\programs\celltypes\workgroups\mousecelltypes\SarahWB\datasets\human_exc\analysis\20251010_seaad\ephys_subthresh'
os.makedirs(subthresh_out_dir, exist_ok=True)
```

Original layer groups

```python
layers = ['L2/3', 'L4', 'L5', 'L6']
subclass_order_dict = {
    'L2/3' : ['L2/3 IT'], 
    'L4' : ['L4 IT'], 
    'L5' : ['L5 IT', 'L5/6 NP', 'L5 ET'], 
    'L6' : ['L6 CT', 'L6 IT', 'L6 IT Car3', 'L6b']
}

for layer in layers:
    print(layer)

    # process layer data 
    layer_data = data_df[data_df[subclass_col].isin(subclass_order_dict[layer])]
    subthresh_cells_dict = process_ephys_subthresh_data(layer_data)

    # save processed data
    np.save(os.path.join(subthresh_out_dir, 'subthreshold_data_{}.npy'.format(layer.replace(' ', '_').replace('/', '-'))), subthresh_cells_dict)

#all cells
print('subclass')

# process layer data 
subthresh_cells_dict = process_ephys_subthresh_data(data_df)

# save processed data
np.save(os.path.join(subthresh_out_dir, 'subthreshold_data_subclass.npy'), subthresh_cells_dict)

data_df.to_csv(os.path.join(subthresh_out_dir, 'subthreshold_metadata.csv'), index=False)
```

Split L2, L3

```python
#Get L2 L3 ttype split
with open(r"\\allen\programs\celltypes\workgroups\mousecelltypes\SarahWB\datasets\human_exc\data\formatting_dicts\ttype_layer_dict.json", "r") as f:
    ttype_to_layer_dict = json.load(f)
    
for layer in ['L3']: #['L2', 'L3']:
    print(layer)
    ttypes = [t for t in ttype_to_layer_dict.keys() if ttype_to_layer_dict[t] == layer.replace('L', 'Layer ')]
    print(ttypes)

    # process layer data 
    layer_data = data_df[data_df[cluster_col].isin(ttypes)]
    subthresh_cells_dict = process_ephys_subthresh_data(layer_data)

    # save processed data
    np.save(os.path.join(subthresh_out_dir, 'subthreshold_data_{}.npy'.format(layer.replace(' ', '_').replace('/', '-'))), subthresh_cells_dict)
```

# Process rheobase data

```python
rheobase_out_dir = r'\\allen\programs\celltypes\workgroups\mousecelltypes\SarahWB\datasets\human_exc\analysis\20251010_seaad\ephys_rheo'
os.makedirs(rheobase_out_dir, exist_ok=True)
```

Original layer groups

```python
layers = ['L2/3', 'L4', 'L5', 'L6']
subclass_order_dict = {
    'L2/3' : ['L2/3 IT'], 
    'L4' : ['L4 IT'], 
    'L5' : ['L5 IT', 'L5/6 NP', 'L5 ET'], 
    'L6' : ['L6 CT', 'L6 IT', 'L6 IT Car3', 'L6b']
}

for layer in layers:
    print(layer)

    # process subclass data 
    layer_cells = list(data_df[data_df[subclass_col].isin(subclass_order_dict[layer])].cell_specimen_name)
    rheobase_sweep_data, rheobase_issues = process_ephys_rheobase_data(layer_cells)

    # save processed data
    np.save(os.path.join(rheobase_out_dir, 'rheo_spike_data_{}.npy'.format(layer.replace(' ', '_').replace('/', '-'))), rheobase_sweep_data, allow_pickle=True)

    # save list of failed cells
    with open(os.path.join(rheobase_out_dir, 'rheo_issues_{}.txt'.format(layer.replace(' ', '_').replace('/', '-'))), 'w') as f:
        for cell in rheobase_issues:
            f.write(cell + '\n')

#all cells
print('subclass')

# process subclass data 
rheobase_sweep_data, rheobase_issues = process_ephys_rheobase_data(list(data_df.cell_specimen_name))

# save processed data
np.save(os.path.join(rheobase_out_dir, 'rheo_spike_data_subclass.npy'), rheobase_sweep_data, allow_pickle=True)

# save list of failed cells
with open(os.path.join(rheobase_out_dir, 'rheo_issues_subclass.txt'), 'w') as f:
    for cell in rheobase_issues:
        f.write(cell + '\n')

data_df.to_csv(os.path.join(rheobase_out_dir, 'rheo_metadata.csv'), index=False)
```

Split L2, L3

```python
#Get L2 L3 ttype split
with open(r"\\allen\programs\celltypes\workgroups\mousecelltypes\SarahWB\datasets\human_exc\data\formatting_dicts\ttype_layer_dict.json", "r") as f:
    ttype_to_layer_dict = json.load(f)
    
for layer in ['L2', 'L3']:
    print(layer)
    ttypes = [t for t in ttype_to_layer_dict.keys() if ttype_to_layer_dict[t] == layer.replace('L', 'Layer ')]
    print(ttypes)

    layer_cells = list(data_df[data_df[cluster_col].isin(ttypes)].cell_specimen_name)
    rheobase_sweep_data, rheobase_issues = process_ephys_rheobase_data(layer_cells)

    # save processed data
    np.save(os.path.join(rheobase_out_dir, 'rheo_spike_data_{}.npy'.format(layer.replace(' ', '_').replace('/', '-'))), rheobase_sweep_data, allow_pickle=True)

    # save list of failed cells
    with open(os.path.join(rheobase_out_dir, 'rheo_issues_{}.txt'.format(layer.replace(' ', '_').replace('/', '-'))), 'w') as f:
        for cell in rheobase_issues:
            f.write(cell + '\n')
```
