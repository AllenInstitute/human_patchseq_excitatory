```python
import os
import pandas as pd
from neuron_morphology.swc_io import morphology_from_swc
```

```python
layer_aligned_dir = r'\\allen\programs\celltypes\workgroups\mousecelltypes\SarahWB\datasets\human_exc\data\SWC_LayerAligned'

soma_depths = {}
for swc in os.listdir(layer_aligned_dir):
    if swc.endswith('.swc'):
        spid = int(swc.split('.swc',1)[0])
        nrn = morphology_from_swc(os.path.join(layer_aligned_dir, swc))
        soma_depths[spid] = nrn.get_soma()['y']
```

```python
soma_depths_df = pd.DataFrame(list(soma_depths.items()), columns=['specimen_id', 'layer_aligned_depth'])
```
